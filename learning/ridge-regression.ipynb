## Prepare

Change price from `object` to `float`

```python
auto_data['price'].describe()
```

```python
auto_data['price'] = pd.to_numeric(auto_data['price'], errors='coerce')
```

Remove column that is not useful in predicting price

```python
auto_data = auto_data.drop('normalized-losses', axis=1)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data.info()
```

Convert `horsepower` to numeric

```python
auto_data['horsepower'].unique()
```

```python
auto_data['horsepower'].describe()
```

```python
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'], errors='coerce')
```

```python
auto_data['horsepower'].describe()
```

Convert `cylinders` from categorical to numeric

```python
auto_data['num-of-cylinders'].describe()
```

```python
auto_data['num-of-cylinders'].unique()
```

```python
cylinders_dict = {
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'eight': 8,
    'twelve': 12
}
auto_data['num-of-cylinders'].replace(cylinders_dict, inplace=True)
auto_data.head()
```

```python
auto_data['num-of-cylinders'].describe()
```

Convert categorical values into one-hot representations

```python
auto_data = pd.get_dummies(auto_data,
                            columns=['make',
                                    'fuel-type',
                                    'aspiration',
                                    'num-of-doors',
                                    'body-style',
                                    'drive-wheels',
                                    'engine-location',
                                    'engine-type',
                                    'fuel-system'])
auto_data.head()
```

Drop rows with null values

```python
auto_data = auto_data.dropna()
```

Check for any null values

```python
auto_data[auto_data.isnull().any(axis=1)]
```

## Train

```python
from sklearn.model_selection import train_test_split

X = auto_data.drop('price', axis=1)

Y = auto_data['price']

X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state=0)
```

When updating `alpha` observe the impact it has on the coefficients.

```python
ridge_model = Ridge(alpha=0.5, normalize=True)
ridge_model.fit(X_train, Y_train)
```

## Evaluate

```python
# get the R^2 score
ridge_model.score(X_train, Y_train)
```

Ridge regression is a regularised regression model which penalises complex coefficients.
The high coefficient values are much lower than in Lasso and simple linear regression.
The penalty function has constrained the complexity of these coefficients.

```python
predictors = X_train.columns
coef = pd.Series(ridge_model.coef_, predictors).sort_values()

print(coef)
```

```python
y_predict = ridge_model.predict(x_test)
```

R^2 much improved compared to linear regression
